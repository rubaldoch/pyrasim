```python
from pyrasim.signals import Sinusoid, ComplexSinusoid, SineWaveform, CosineWaveform, ComplexSinusoidWaveform
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_1 = Sinusoid(2, 1/4)
x_2 = Sinusoid(3, 1/4, -np.pi/2)

x = x_1+x_2

x_1.plot(1000)
x_2.plot(1000)
x.plot(1000)
```

```python
sineWaveform = SineWaveform(frequency=5, time_start=0, duration=3, sample_rate=1000)
cosineWaveform = CosineWaveform(frequency=-7, time_start=0, sample_rate=1000)

sineWaveform.plot()

sineWaveform.update_sample_rate(1000)
sineWaveform.plot()
cosineWaveform.plot()

waveform = sineWaveform + cosineWaveform
waveform.plot()

waveform = cosineWaveform + sineWaveform
waveform.plot()
```

```python
xc_1 = ComplexSinusoid(2, 1/8, np.pi/3)
xc_2 = ComplexSinusoid(2, -1/8, -np.pi/3)
xc_1.plot()
xc_2.plot()
```

```python
complexSinusoidWaveform = ComplexSinusoidWaveform(sample_rate=1000)
complexSinusoidWaveform2 = ComplexSinusoidWaveform(frequency=20, sample_rate=1000)
complexSinusoidWaveform.plot()
complexSinusoidWaveform2.plot()
resultado = complexSinusoidWaveform + complexSinusoidWaveform2
resultado.plot()
```
